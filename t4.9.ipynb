```python
a = int(input())
while a!=1:
    if a % 2==0:
        t = a / 2
        print("%d/2 = %d"%(a,t))
        a = t
    else:
        t = a * 3 +1;
        print("%d*3+1=%d"% (a, t))
        a = t
print("End")
```

```python
a = int(input())
num = [1]*a
for i in range(1,a):
    num[i] = num[i-1] * (i+1)
print(sum(num)) 
```

```python
n = int(input())
ans = 0;
for i in range(1,n+1):
    ans = ans + (1.0/i)*((-1)**(i+1))
print("%6f"%ans)
```

```python
n = int(input())
num = [1]*n
for i in range(1,n):
    num[i] = num[i-1] * (i+1)
ans = 1
for i in range(n):
    ans = ans + 1/num[i]
print("%.10f"%ans)
```

```python
import math
n = int(input())
for i in range(2, int(math.sqrt(n))+1):
    while n % i == 0:
        print(i, end=' ')
        n = n // i
if n > 2:
    print(n)
```

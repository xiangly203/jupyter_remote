```python
from time import time
import numba
from numba import jit
import numpy as np
t1 = time()

@jit(nopython=True)
def ess(n):  # 求2到n范围内所有素数
    es = [2]
    Td = [True] * (n + 1)
    for i in range(3, n + 1, 2):
        if Td[i]:
            es.append(i)
            for j in range(i ** 2, n + 1, 2 * i):
                Td[j] = False
    return es


n = 10 ** 9
es = ess(n)
ff = f"从2到10亿范围内素数共{len(es)}个"

t2 = time()
tt = t2 - t1
print("\n", ff, f"\n共耗时{tt}秒")
```
